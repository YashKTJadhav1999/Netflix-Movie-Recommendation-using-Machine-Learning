**Installing Libraries**

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Importing Dataset**

```python
df = pd.read_csv("/content/Netflix_Recommendation_System.csv")
```

```python
df
```

```python
movies = []

for i in range(0,7467):
  movies.append([str(df.values[i,j]) for j in range(0,22)])
```

**Apriori training dataset**

```python
from apyori import apriori
```

```python
netflix_ai = apriori(transactions = movies, min_support = 0.002, min_confidence=0.2, min_lift=3, min_length=2, max_length=2)
```

Visualising the Results

**Raw Results**

```python
results = list(netflix_ai)
results
```

**Results displayed in a Better format**

```python
def inspect(results):
  movie1 = [tuple(result[2][0][0])[0] for result in results]
  movie2 = [tuple(result[2][0][1])[0] for result in results]
  supports = [result[1] for result in results]
  return list(zip(movie1,movie2,supports))

DF_AI = pd.DataFrame(inspect(results),columns=['Movie1','Movie2','Support'])
DF_AI
```

**Top 10 Movie Relationship based on largest Support**

```python
DF_AI.nlargest(n=10,columns='Support')
```
